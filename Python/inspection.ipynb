```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML

# 1. Define file paths using raw strings for Windows
file_paths = {
    "Ads": r"D:\Data_Science\Projects\Portfolio_Projects\5\Dataset\Raw\ads.csv",
    "Campaigns": r"D:\Data_Science\Projects\Portfolio_Projects\5\Dataset\Raw\campaigns.csv",
    "Users": r"D:\Data_Science\Projects\Portfolio_Projects\5\Dataset\Raw\users.csv",
    "Ad Events": r"D:\Data_Science\Projects\Portfolio_Projects\5\Dataset\Raw\ad_events.csv"
}

# 2. Load the datasets into a dictionary
datasets = {}
for name, path in file_paths.items():
    try:
        datasets[name] = pd.read_csv(path)
        print(f"✅ Successfully loaded: {name}")
    except Exception as e:
        print(f"❌ Error loading {name}: {e}")

# 3. Extended comprehensive inspection function
def inspect_data(df, name):
    print(f"\n{'='*60}")
    print(f"📊 DATASET: {name.upper()}")
    print(f"{'='*60}")
    
    # Overview
    print("\n--- 1. Overview ---")
    print(f"Rows: {df.shape[0]:,}")
    print(f"Columns: {df.shape[1]:,}")
    print(f"Memory Usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    print(f"Total Duplicated Rows: {df.duplicated().sum():,}")
    
    # First 3 Rows
    print("\n--- 2. First 3 Rows ---")
    display(df.head(3))
    
    # Missing Values (Count & Percentage)
    print("\n--- 3. Missing Values ---")
    missing = df.isnull().sum()
    missing_pct = (missing / len(df)) * 100
    missing_df = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
    missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values(by='Missing %', ascending=False)
    
    if not missing_df.empty:
        display(missing_df)
    else:
        print("No missing values found. ✅")
    
    # Summary Statistics (Numerical)
    print("\n--- 4. Numerical Summary ---")
    num_cols = df.select_dtypes(include=[np.number]).columns
    if len(num_cols) > 0:
        display(df.describe().T)
    else:
        print("No numerical columns found.")

    # Summary Statistics (Categorical)
    print("\n--- 5. Categorical Summary ---")
    cat_cols = df.select_dtypes(include=['object', 'category', 'string']).columns
    if len(cat_cols) > 0:
        display(df.describe(include=['object', 'category', 'string']).T)
    else:
        print("No categorical columns found.")
        
    # Unique Values Preview (Max 15)
    print("\n--- 6. Unique Values Preview (Max 15) ---")
    for col in df.columns:
        n_unique = df[col].nunique()
        unique_vals = df[col].unique()
        
        # Limit to first 15 unique values for cleaner output
        if len(unique_vals) > 15:
            displayed_vals = list(unique_vals[:15]) + ["... (truncated)"]
        else:
            displayed_vals = list(unique_vals)
            
        print(f"• {col} ({n_unique} unique): {displayed_vals}")

# 4. Execute inspection for each loaded dataset
for name, df in datasets.items():
    inspect_data(df, name) 
```
